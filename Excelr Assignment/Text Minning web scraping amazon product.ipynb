# calculating sentiment value for review

```python
import pandas as pd
import spacy
affin=pd.read_csv(r'C:\Users\Karan\Downloads\affin.csv.csv',sep=',',encoding='latin-1')
affin.shape
```

```python
affin.head()
```

```python
affin_scores=affin.set_index('word')['value'].to_dict()
```

```python
nlp=spacy.load('en_core_web_sm')
sentiment_lexicon=affin_scores


def calculate_sentiment(text: str=None):
    sent_score=0
    if text:
        sentence=nlp(text)
        for word in sentence:
            sent_score +=sentiment_lexicon.get(word.lemma_,0)
        return sent_score
```

```python
calculate_sentiment(text='nice')
```

```python
df['sentiment_value']=df['Reviews'].apply(calculate_sentiment)
```

```python
def sentiment_cat(row):
    
    '''This function returns sentiment category based on values in sentiment_value column'''
    
    if row['sentiment_value'] < 0:
        val = 'Negative'
    elif row['sentiment_value'] == 0: 
        val = 'Neutral'
    elif row['sentiment_value'] > 0:
        val = 'Positive'
    else:
        val = -1
    return val
```

```python
df['sentiment_category'] = df.apply(sentiment_cat, axis=1)
df.head()
```
